```python
import numpy as np
import pandas as pd
import multiprocessing
import matplotlib.pyplot as plt
import seaborn as sns

def process_data(chunk, results):

    # 1. Menghitung jumlah data dalam setiap chunk
    count = len(chunk)
    results['Data count'] = count

    # 2. Mencari usia tertua dalam setiap chunk
    max_age = chunk['age'].max()
    results['Max Age'] = max_age

    # 3. Mencari nilai premi rata-rata dalam setiap chunk
    avg_premium = chunk['charges'].mean()
    results['Average Premium'] = avg_premium

    # 4. Rata-rata umur
    average_age = chunk['age'].mean()
    results['Average Age'] = average_age

if __name__ == '__main__':
    df = pd.read_csv('insurance.csv')

    num_chunks = 4
    chunks = np.array_split(df, num_chunks)

    manager = multiprocessing.Manager()
    results = manager.dict()

    processes = []
    for chunk in chunks:
        process = multiprocessing.Process(target=process_data, args=(chunk, results))
        processes.append(process)

    for process in processes:
        process.start()

    for process in processes:
        process.join()

    print("Data count:", results['Data count'])
    print("Max Age:", results['Max Age'])
    print("Average Premium:", results['Average Premium'])
    print("Average Age:", results['Average Age'])

# Visualisasi distribusi umur
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], kde=True)
plt.title('Distribusi Umur')
plt.xlabel('Umur')
plt.ylabel('Count')
plt.show()

# Visualisasi hubungan antara umur dan biaya premi
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='charges', data=df)
plt.title('hubungan antara umur dan biaya premi')
plt.xlabel('Umur')
plt.ylabel('Biaya Premi')
plt.show()

# Visualisasi barplot untuk biaya premi berdasarkan status merokok
plt.figure(figsize=(8, 6))
sns.barplot(x='smoker', y='charges', data=df)
plt.title('Biaya Premi Berdasarkan Status Merokok')
plt.xlabel('Perokok')
plt.ylabel('Biaya Premi')
plt.show()

# Visualisasi hubungan antara indeks massa tubuh (bmi) dan biaya premi
plt.figure(figsize=(8, 6))
sns.scatterplot(x='bmi', y='charges', data=df)
plt.title('Hubungan Antara Indeks Massa Tubuh (bmi) dan Biaya Premi')
plt.xlabel('Indeks Massa Tubuh')
plt.ylabel('Biaya Premi')
plt.show()
```
